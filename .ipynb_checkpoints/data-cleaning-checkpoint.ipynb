```python
import pandas as pd
```

```python
# Store countries.csv from URL in pandas data frame
url = 'https://raw.githubusercontent.com/edlich/eternalrepo/master/DS-WAHLFACH/dsm-beuth-edl-demodata-dirty.csv'
df = pd.read_csv(url, index_col=0)
```

Display dataframe to inspect content in situ

```python
display(df)
```

From visually inspection we see that we have:
<ol>
    <li>NaN indices</li>
    <li>NaN values</li>
    <li>Negative ages and string as age</li>
    <li>Doube rows</li>
</ol>

1. Drop all NaN values and check for NaN values

```python
df = df.dropna()
```

```python
df.isnull().values.any()
```

2. Recreate indices and check for NaN indices

```python
df = df.reset_index(drop=True)
```

```python
df.loc[df.index.isnull()]
```

3. Remove all rows that contain not numbers or negative age (we do not know if 72 woul be the true age, or if it's a total typo) 

```python
df = df[df.age != "old"]
```

```python
df = df[df.age.astype(int) >= 0]
```

4. Remove all double rows and keep one

```python
df = df.drop_duplicates()
```

Finaly display cleaned dataframe

```python
display(df)
```
